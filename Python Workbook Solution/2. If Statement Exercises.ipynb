The roots of a quadratic function can be found by finding the values of  $$x$$  that satisfy the quadratic equation  $$ax^2 + bx + c = 0$$ . A quadratic function may have 0, 1 or 2 real roots. These roots can be computed using the quadratic formula, shown below:
 $$\begin{aligned} root = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a} \end{aligned}$$ 
The portion of the expression under the square root sign is called the discriminant. If the discriminant is negative then the quadratic equation does not have any real roots. If the discriminant is 0, then the equation has one real root. Otherwise the equation has two real roots, and the expression must be evaluated twice, once using a plus sign, and once using a minus sign, when computing the numerator.

```python
from math import sqrt
from fractions import Fraction as f

a = int(input("Enter the value of a: "))
b = int(input("Enter the value of b: "))
c = int(input("Enter the value of c: "))

if a == 0:
    print("Incorrect equation the value of 'a' cannot be zero.")
else:
    disc = b**2 - 4*a*c
    root1 = f((-b + sqrt(abs(disc)))/(2*a)).limit_denominator()
    root2 = f((-b - sqrt(abs(disc)))/(2*a)).limit_denominator()
    
    if disc > 0 :
        print(f"This equation have two real Root, and the Root is {root1} and {root2}")
    elif disc == 0:
        print(f"This equation have real Root, and the Roots are equal which is {root1}")
    else:
        print(f"This equation doesn't have any real Roots")   
```

Of the remaining years, any year that is divisible by 100 is not a leap year.
Of the remaining years, any year that is divisible by 4 is a leap year.
All other years are not leap years.
